```python


df = pd.read_csv(f"ciao.csv", header=None).set_index([1, 2, 3, 4, 0]).sort_index()


df = df.unstack()

df.columns = [ c0[1] for c0 in df.columns ]

cc = df.columns

mx = []
for i in range(len(cc)):
    for j in range(len(cc)):
        if cc[i] != cc[j]:
#             print(cc[i], cc[j], (df[cc[i]] == df[cc[j]]).all())
            mx.append([cc[i], cc[j], (df[cc[i]] == df[cc[j]]).all()])
            
print(pd.DataFrame(mx).pivot(index=0, columns=1, values=2))
```
