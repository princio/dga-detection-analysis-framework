```python
import json
import os

dgarchive_bigfields = ['guesses', 'hits', 'misses']

def parse_response(file):
    data = {}
    with open(f'tmp/{file}.success') as fp:
        res = json.load(fp)
        for macro in res:
            data[macro] = { f'{item['domain']}': item for item in res[macro] }
    return data

files = list(set([ file[:file.find('.')] for file in list(os.walk('tmp'))[0][2] ]))

DF_s = []
for file in files:
    if os.path.exists(f'tmp/{file}.success'):
        further_values = []
        df = pd.read_csv(f'tmp/{file}.list.csv', index_col=0).reset_index(drop=True)

        data = parse_response(file)
        
        for _, row in df.iterrows():
            further_values.append([ data[tmp][row['dn']] if row['dn'] in data[tmp] else None for tmp in dgarchive_bigfields ])
            pass
        DF_s.append(pd.concat([df, pd.DataFrame(further_values, columns=dgarchive_bigfields)], axis=1))
        pass
    pass

for l in dgarchive_bigfields:
    DF = pd.concat(DF_s)
    display(DF[~DF[l].isna()].head())
```
